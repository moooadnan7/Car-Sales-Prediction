```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
cars_data=pd.read_csv('/content/car_purchasing.csv',encoding="ISO-8859-1")
cars_data
```

```python
cars_data.describe()
```

```python
print("Number of customers :",cars_data['customer name'].count())
```

```python
print("the maximum age is :",cars_data['age'].max())
print("the minimum age is :",cars_data['age'].min())
print("the average age is :",cars_data['age'].mean())
```

```python
print("the purchase amount is :",cars_data['car purchase amount'].max())
print("the purchase amount is :",cars_data['car purchase amount'].min())
print("the purchase amount is :",cars_data['car purchase amount'].mean())
```

```python
cars_data.columns
```

```python
cars_data.drop(['customer name','customer e-mail','country','gender'],axis=1,inplace=True)
cars_data
```

```python
x=cars_data.drop('car purchase amount',axis=1)
y=cars_data['car purchase amount']
```

```python
st=MinMaxScaler()
x=st.fit_transform(x)
y=st.fit_transform(y.values.reshape(-1,1))
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Sequential()
model.add(Dense(20, input_dim=4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_squared_error'])
```

```python
history=model.fit(x_train, y_train, epochs=50,validation_split=0.2 )
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

R2 = r2_score(y_test, y_pred)
print("R2 Score=",R2 )
```
